```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")  # gaya latar grid lembut untuk plot berikutnya
df = sns.load_dataset("titanic")
df.head()  # cuplikan baris untuk cek kolom
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))  # dua sumbu sejajar
sns.histplot(df["age"].dropna(), kde=True, ax=ax[0])  # kde: estimasi kepadatan
ax[0].set_title("Distribusi umur penumpang")
ax[0].set_xlabel("age")
sns.scatterplot(
    data=df, x="age", y="fare", hue="survived", alpha=0.65, ax=ax[1]  # hue: warna per kelas survived
)
ax[1].set_title("Umur vs tarif (hue: survived)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x="class", y="fare")  # perbandingan distribusi fare per kelas kabin
plt.title("Sebaran fare per class")
plt.xlabel("class")
plt.ylabel("fare")
plt.tight_layout()
plt.show()
```

```python
num = df[["survived", "pclass", "age", "sibsp", "parch", "fare"]].dropna()  # baris lengkap untuk matriks korelasi
plt.figure(figsize=(7, 5))
sns.heatmap(num.corr(), annot=True, fmt=".2f", cmap="vlag", center=0)  # Pearson; center=0: diverging colormap
plt.title("Korelasi antar variabel numerik")
plt.tight_layout()
plt.show()
```

```python
cols = ["age", "fare", "pclass"]
sns.pairplot(
    df[cols + ["survived"]].dropna(),
    hue="survived",
    corner=True,  # hanya segitiga bawah
    plot_kws={"alpha": 0.6},
)
plt.show()
```
